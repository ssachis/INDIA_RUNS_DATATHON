```python
import numpy as np

proficiency_weights = {
    "beginner": 25,
    "intermediate": 50,
    "advanced": 75,
    "expert": 90
}

def calculate_skill_score(row):
    prof = proficiency_weights.get(
        str(row["proficiency"]).lower(),
        50
    )

    # logarithmic endorsement scaling
    endorsement_score = min(
        np.log1p(row["endorsements"]) / np.log1p(100) * 100,
        100
    )

    # experience saturation at 5 years
    duration_score = min(
        np.log1p(row["duration_months"]) / np.log1p(60) * 100,
        100
    )

    score = (
        0.50 * prof +
        0.20 * endorsement_score +
        0.30 * duration_score
    )

    return round(score, 1)
```
